```python
# The second package we want to look at is matplotlib.pyplot (The pyplot module from the matplotlib)
# This module is all about visualizing data

# First we generate a dataset as we did before using numpy
# Also we want to introduce abbreviations for modules here:
# Using the keyword 'as' we can define an alternative name for a module to be used instead of the full name
# Common uses are the following:

import numpy as np

data = np.loadtxt(fname='data/inflammation-01.csv', delimiter=',')

import matplotlib.pyplot as plt
image = plt.imshow(data)
plt.show()
```

```python
# We can combine numpy and matplotlib to get plot the average inflamation for each days
ave_inflammation = np.mean(data, axis=0)
ave_plot = plt.plot(ave_inflammation)
plt.show()

max_plot = plt.plot(np.max(data, axis=0))
plt.show()

min_plot = plt.plot(np.min(data, axis=0))
plt.show()

# The other direction is no useful information as the average of all days for each patient
# has no meaning. One can however see how the axes behave
min_plot = plt.plot(np.mean(data, axis=1))
plt.show()
```

```python
data = np.loadtxt(fname='data/inflammation-01.csv', delimiter=',')

# We can create multiple plots in the same figure using plt.subplots
# The pair of numbers sets rows, columns of figures side by side
fig, ax = plt.subplots(3,1,figsize=(4.0, 10.0))
#fig, ax = plt.subplots(1,3,figsize=(4.0, 10.0))

# The ax object is a numpy array of plot axes (object type)
ax[0].set_ylabel('average')
ax[0].plot(np.mean(data, axis=0))

ax[1].set_ylabel('max')
ax[1].plot(np.max(data, axis=0))

ax[2].set_ylabel('min')
ax[2].plot(np.min(data, axis=0))

plt.savefig('inflammation.png')
plt.show()
```

```python
# Task: Create a plot showing the standard deviation across the patients 
std_plot = plt.plot(np.std(data, axis=0))
plt.show()
```

```python
# Task: Create a function which takes a string filename and 
# plots the average, maximum inflammation for each patient
# Advanced portion: Set an empty linestyle and markers to an x (help(plt.plot))

def plot_indicators(filename):
    data = np.loadtxt(fname=filename, delimiter=',')

    fig, ax = plt.subplots(2,1,figsize=(10.0, 3.0))

    ax[0].set_ylabel('average')
    ax[0].plot(np.mean(data, axis=1), linestyle='', marker='x')

    ax[1].set_ylabel('max')
    ax[1].plot(np.max(data, axis=1), linestyle='', marker='x')

    plt.show()
        
plot_indicators('data/inflammation-01.csv')
```

```python
# Reading multiple datasets
# We can create a list of the names of the datasets we want to iterate over
# With the glob function
import glob

print(glob.glob('data/inflammation*.csv'))
```

```python
# Using the glob we can automatically find all files we want to iterate over
# Task: Use the glob to read inflammation-01, inflammation-02 and inflammation-03 and
# use a for loop plot the maximum and standard deviation for each patient

filenames = sorted(glob.glob('data/inflammation*.csv'))
filenames = filenames[0:3]
for filename in filenames:
    print(filename)
    plot_indicators(filename)
```
