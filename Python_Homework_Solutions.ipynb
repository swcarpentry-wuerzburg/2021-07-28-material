```python
# 1
def factorial(n):
    result = 1
    for i in range(1,n+1):
        result = result*i
    return result

print(factorial(0))
print(factorial(1))
print(factorial(2))
print(factorial(3))
print(factorial(4))
print(factorial(5))
```

```python
# 2, one possible solution
def add_two_lists(a,b):
    n_a = len(a)
    n_b = len(b)
    result = []
    if n_a > n_b:
        result = list(a)
        for i in range(n_b):
            result[i] = result[i]+b[i]
    else:
        result = list(b)
        for i in range(n_a):
            result[i] = result[i]+a[i]

    return result

list1 = [1,2,3,4, 5]
list2 = [2,4,6,8,10,12]
print(add_two_lists(list1,list2))
print(add_two_lists(list2,list1))
```

```python
#3, one possible solution
def slow_and_simple_sort(inp):
    for i in range(len(inp)):
        # Let us first find the smallest element in the range
        # between i and len(inp)
        smallest_element = inp[i]
        smallest_position = i
        for j in range(i,len(inp)):
            if inp[j] < smallest_element:
                smallest_element = inp[j]
                smallest_position = j
        
        # Now we swap the smallest element into the i-th position
        was_at_i = inp[i]
        was_at_smallest_position = inp[smallest_position]
        inp[i] = was_at_smallest_position
        inp[smallest_position] = was_at_i
        
    return inp
slow_and_simple_sort([5,4,3,0,6.6,2,1])
        
```

```python
#4 
def split_a_list(inp):
    smaller10 = []
    greater10 = []
    for number in inp:
        if number < 10:
            smaller10.append(number)
        elif number > 10:
            greater10.append(number)
        else:
            print("10 does not get added")
            
    return smaller10, greater10

s,g = split_a_list([10,11,2,20,4,-2])
print(s)
print(g)
```

```python
#5 
def get_first_character(list_of_words):
    result = []
    for word in list_of_words:
        result.append(word[0])
    return result
def get_last_character(list_of_words):
    result = []
    for word in list_of_words:
        result.append(word[-1])
    return result
print(get_first_character(["tim", "anna", "hund", "katze", "maus"]))
print(get_last_character(["tim", "anna", "hund", "katze", "maus"]))
```
