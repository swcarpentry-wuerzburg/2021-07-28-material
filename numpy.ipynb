```python
# Review of yesterday:

# We can store lists into variables
primes = [2,3,5,7,11,13]

# We can print it
print("The complete list",primes)
for prime in primes:
    print("a prime",prime)
print("")

# We can also use the range and len functions to get an index
for i in range(0,len(primes)):
    print("Prime number",i,"is",primes[i])
print("")

# We can also write a function:
def add_elements_of_list_smaller_than_10(input_list):
    result = 0
    for element in input_list:
        if element < 10:
            result = result + element
        else:
            print(element,"is larger than 10 and does not get added")
    return result

# Let us call it and print the result
sum_of_primes_smaller_10 = add_elements_of_list_smaller_than_10(primes)
print("")
print("The sum of all primes < 10 is",sum_of_primes_smaller_10)
```

```python
# Having now understood some basic concepts of python we want
# to use build-in methods and packages to make these tasks easier
# For this we shall introduce numpy and matplotlib, starting with numpy

# In python we can load packages using the 'import' keyword
import numpy
```

```python
# We can acess functions and elements of this package with the . operator
print( numpy.pi )
print( numpy.sqrt(9) )
```

```python
# To make typing faster it is common to use abbreviations for the libraries, we do this with the as statement
import numpy as np

# Now we can just type, i.e. refer to numpy by using np
print( np.pi )
print( np.sqrt(9) )
```

```python
# To work with data in numpy, we usually use np.array() objects, they are very similar to lists
a_list = primes
an_array = np.array(primes)

for prime in a_list:
    print("a prime",prime)
print("")
print("works the same for the numpy array")
print("")
for prime in an_array:
    print("a prime",prime)
```

```python
# We can load the csv data with the loadtxt function from the numpy library
# This function takes a number of arguments, we can specify which we are setting with the 'name=' argument
# The others are set to default values

data = np.loadtxt(fname='data/inflammation-01.csv', delimiter=',')

# Let us analyze the data we have loaded
print(data)
print(type(data))

# The numpy.ndarray class has built-in function to get the content type and size
# Unlike a list, a numpy array only ever contains one type of data
print(data.dtype)
print(data.shape)

# Let us unpack data.shape
dimensions = data.shape
print(dimensions)
n_rows = dimensions[0]
n_cols = dimensions[1]
print("number of rows",n_rows)
print("number of columns",n_cols)
```

```python
# We can access the numpy array's data in the same way we would access a list,
# Here we also introduce the idea of 2D data,
# Therefore we need a pair of indices seperated by a comma
print('first row in data:', data[0])
print('first element of first row in data:', data[0][0])

print('first element of first row in data in a more compact way:', data[0, 0])
```

```python
# Maybe we don't want to print a line but a column, we could do
n_rows = data.shape[0]
for ii in range(0,n_rows):
    print(data[ii,4])
# printing the content of the "fourth" column
```

```python
# But there is an easier way:
print(data[:,4])
```

```python
# But we want to introduce a new concept: slicing (also works on lists)
# Here the slice '0:4' means start at 0 and go up to - not including - 4
print(data[0:4, 0:10])

# We can ommit the upper or lower bound, that will mean start/beginning
# If we use just ':' that means all elements
small = data[:3, 36:]
print('small is:')
print(small)

# We can get one row of the data with '[0,:]'
print(small[0,:])

# Or alternatively we can get one column of the data with '[:,0]'
print(small[:,0])
```

```python
# Remember that strings share access concepts with lists and arrays? We can also slice strings
# Task: Given a string get the last element 
test_string = 'carpentry'

# Task Write a function which removes the first and last element of a string (using slicing)
def middle_part(input_string):
    print(input_string[1:-1])
    return input_string[1:-1]

# Task: Write a function which adds all elements of a numpy array (element-wise)
def add_rows(input_array):
    result = 0
    for row in input_array:
        for element in row:
            result = result + element
    return result

def add_rows_adv(input_array):
    return np.sum(input_array)

middle_part(test_string)
print(add_rows(data))
print(add_rows_adv(data))
```

```python
# Numpy has many functions to analyze arrays for example we can compute mean values
print("Mean",np.mean(data))

# We can also compute maxima, minima and standard deviations of the dataset
# Note that we can assign multiple values in one line with a comma seperation on both sides
maxval, minval, stdval = np.max(data), np.min(data), np.std(data)

print('maximum inflammation:', maxval)
print('minimum inflammation:', minval)
print('standard deviation:', stdval)
```

```python
# We can now extend our repertoir of functions to include the most common numpy functions:

# numpy.arange() is the extension of the range() function to non-integer numbers
# It however returns a numpy.ndarray not a list
print(numpy.arange(5,8,2))

#? print(range(0.5,4.8,0.3))
print(numpy.arange(0.5,4.8,0.5))

# The stop float is NOT included in the array 
print(numpy.arange(0.5,5.0,0.5))
print(numpy.arange(0.5,5.1,0.5))

print(type(numpy.arange(0,4,2)))

# numpy.linspace instead provides N evenly spaced numbers in a given intervall
print(numpy.linspace(1,2,10))
print(len(numpy.linspace(1,2,10)))
```

```python
# Task: Write a function which computes the mean of all natural numbers
# between two given integers, Hint: This can be done without a for loop using numpy
def mean_between(a,b):
    return numpy.mean(numpy.arange(a+1,b,1))

print(mean_between(2345, 6723))
```

```python
# Task: Write a function that adds two arrays of equal length elementwise
# and returns a numpy array

def array_add(a,b):
    result = []
    for i in range(len(a)):
        result.append(a[i]+b[i])
    return numpy.array(result)

a = numpy.linspace(2,4,10)
b = numpy.linspace(1,5,10)

print(array_add(a,b))
print(array_add(a,b) == a+b)

# Talk about numpy array addition and multiplication 
```

```python
# Summary

# Packages are imported like so, the as keyword allows us to pick an alias that we want, some are THE convention
import numpy as np

# Data often has at least a two-dimensional shape is is nicely represented with a 2d numpy array
# e.g. inflammation of the wound for each patient tracked and for each day

# We can easily load .csv files into a numpy array
data = np.loadtxt(fname='data/inflammation-01.csv', delimiter=',')

# We can get the data in a row (for a single patient)
print("Data for patient 0",data[0])
print("")
# Or for a single day:
print("Data on day 0",data[:,0])
print("")
# Or for a specific patient at a specific day
print("Data for patient 5 on day 4",data[5,4])
print("")

# We can use numpy functions to compute statistical quantities
print("Mean of all data",np.mean(data))
for row in data:
    print("Mean inflammation for single patient",np.mean(row))

print(data.shape)
print(data.shape[1])
for column_index in range(0,data.shape[1]):
    data_on_a_day = data[:,column_index]
    print("Mean inflammation for day",column_index,":",np.mean(data_on_a_day))
    
# These lists are tedious to look at if we want to identify trends, so after the break we will learn how to plot such data
```
