```python
# The second package we want to look at is matplotlib.pyplot (The pyplot module from the matplotlib)
# This module is all about visualizing data

# First we get a dataset as we did before using numpy
# Common use for the abbreviations are the following:

# This is the minimal code needed for a simple visualization of our data:

import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt("data/inflammation-01.csv",delimiter=",")

plt.imshow(data)
plt.show()
```

```python
# If we are interested in detailed plots of rows, or columns of the data we may use
plt.plot(data[0])
plt.show()
plt.plot(data[10])
plt.show()
plt.plot(data[:,10])
plt.show()
```

```python
# notice how plt.show() seperates the plots above but allows us to stack the information here
plt.plot(data[0])
plt.plot(data[1])
plt.plot(data[10])
plt.show()
```

```python
# We can also color everything the way we like
plt.plot(data[0],color="black")
plt.plot(data[1],color="blue")
plt.plot(data[10],color="green")
plt.show()
```

```python
# and add all the relevant information to actually make the plot useful
plt.plot(data[0],color="black", label="Patient 0")
plt.plot(data[1],color="blue",  label="Patient 1")
plt.plot(data[10],color="green",label="Patient 10")
plt.xlabel("Day")
plt.ylabel("inflammation")
plt.legend()
plt.show()
```

```python
# this also works with the default colors
plt.plot(data[0], label="Patient 0")
plt.plot(data[1], label="Patient 1")
plt.plot(data[10],label="Patient 10")
plt.xlabel("Day")
plt.ylabel("inflammation")
plt.legend()
plt.show()
```

```python
# remember how , allows us to store objects into multiple variables at the same time
# a, b = 2, 3
# print(a)
# print(b)

# we can use it to split the output of plt.subplots() into the an object to handle the complete figure (fig)
# and an object that handles the seperate subfigures (ax)
fig, ax = plt.subplots(1,3,figsize=(12,4))

# since we created a plot with one row and three columns the ax object is 3 elements long
print("number of elements in the ax object =",len(ax))

# We can acess them with [] brackets like an array, the functions for labeling are different !
ax[0].set_xlabel("Day")
ax[0].plot(data[0])

# The order does not matter
ax[1].plot(data[1])
ax[1].set_xlabel("Day")

ax[2].plot(data[:,20])
ax[2].set_xlabel("Patient")
# we can also set a title ;-)

ax[2].set_title("Inflammation on day 20")

plt.show()
```

```python
# Clearly we can automate all of this so we don't have to label everything by hand
fig, ax = plt.subplots(4,1,figsize=(3,8))
for i in range(len(ax)):
    ax[i].plot(data[i])
    ax[i].set_xlabel("Day")
    ax[i].set_ylabel("Inflammation")
    ax[i].set_title("Data for patient "+str(i) )
plt.tight_layout()
plt.show()
```

```python
# Exercise
# create a function that plots the data of a given
# filename for the first 3 patients into a single plot
# make a legend and label the axes

def plot_first_three(filename):
    data = np.loadtxt(filename,delimiter=",")
    
    fig, ax  = plt.subplots(1,1,figsize=(4,3))
    for i in range(3):
        ax.plot(data[i],label="patient "+str(i))
    ax.legend()
    ax.set_xlabel("Day")
    ax.set_ylabel("Inflammation")
    plt.show()
def plot_first_three_plt(filename):
    data = np.loadtxt(filename,delimiter=",")
    
    for i in range(3):
        plt.plot(data[i],label="patient "+str(i))
    plt.legend()
    plt.xlabel("Day")
    plt.ylabel("Inflammation")
    plt.show()

plot_first_three_plt("data/inflammation-04.csv")
```
